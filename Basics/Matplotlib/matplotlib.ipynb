```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df
```

```python
x = [2017,2018,2019,2020,2021,2022,2023,2024]
y = [43,23,45,56,35,87,34,58]
```

```python
#simple line plot
plt.plot(x,y,marker='o',linestyle='-',color='g',label='INd')
plt.xlabel('year')
plt.ylabel('growth')
plt.title('TensorCodes business analysis')
plt.legend(loc='lower right')
plt.show()
# perfect for timeseries data set
```

```python
plt.bar(df['species'].value_counts().index,df['species'].value_counts().values)
plt.title('iris dataset species frequency distribution')
plt.xlabel('species')
plt.ylabel('frequency')
# plt.legend('lower')
plt.show()
```

```python
plt.pie(df['species'].value_counts().values,labels=df['species'].value_counts().index)
plt.title('tensorcodes')

#we dont need to use these functions for pie chart
# plt.xlabel('tensor')
# plt.ylabel('codes')
# plt.legend('lower right')
```

```python
np.min(df['sepal_length'])
```

```python
bins = [3.5,4,4.5,5,5.5,6,6.5,7,7.5,8]
```

```python
plt.hist(df['sepal_length'],bins,edgecolor='pink',color='black')
plt.title('sepal length distribution')
plt.xlabel('x-axis')
plt.ylabel('sepal_lenght')
plt.legend('lower right')
plt.show()
```

```python
plt.boxplot(df['sepal_length'])
# unlike pandas plot ,we can't plot all colmns boxplot in one plot df.plot.boxplot()
plt.title('sepal length outliers')
plt.xlabel('sepal length')
plt.ylabel('range')
plt.legend('right')
plt.show()
```
