```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,5))
ax = sns.lineplot(df['sepal_width'],)
ax.set(xlabel='index',title='IRIS',)
plt.title("mac") #override the ax's title
```

```python

plt.figure(figsize=(10,5))
sns.barplot(x=df['species'].value_counts().index,y=df['species'].value_counts().values)
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df['sepal_width'],kde=True,color='black') # By kde we can judge this distribution as gossian distribution 
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df,x=df.index,y='sepal_length')
```

```python
plt.figure(figsize=(16,10))
sns.boxplot(df) #by this graph we can say that only sepal_width have outliers
```
