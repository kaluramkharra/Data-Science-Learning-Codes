```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
# we don't need to remenber all the arguments of plots .
# we just need to rembember basic arguments like - x=,y=,data=,palette,hue,color etc.
# go through the official documentation for further customization
```

```python
plt.figure(figsize=(10,5))
sns.set_style("darkgrid")
ax = sns.lineplot(df['sepal_width'],)
ax.set(xlabel='index',title='IRIS',)
plt.title("mac") #override the ax's title
```

```python
plt.figure(figsize=(16,10))
sns.set_style("whitegrid")
sns.lineplot(data = df, y = 'sepal_length',x=df.index,palette='rocket',hue='species')
```

```python

plt.figure(figsize=(10,5))
sns.barplot(x=df['species'].value_counts().index,y=df['species'].value_counts().values,palette='cool')
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df['sepal_width'],kde=True,palette='hot') # By kde we can judge this distribution as gossian distribution 
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df,x=df.index,y='sepal_length',hue=df['species'],palette='magma')
```

```python
plt.figure(figsize=(12,7))
sns.boxplot(df,palette='flare') #by this graph we can say that only sepal_width have outliers
```

```python
plt.figure(figsize=(16,12))
sns.set_style('darkgrid')
sns.heatmap(df.drop(['species'],axis=1))
```
