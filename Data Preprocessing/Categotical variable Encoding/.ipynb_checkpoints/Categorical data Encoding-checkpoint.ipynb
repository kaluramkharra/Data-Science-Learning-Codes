```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips_df = sns.load_dataset('tips')
```

```python
tips_df.head()
```

```python
tips_df.shape
```

```python
pd.get_dummies(tips_df) #One hot encoding for every column
```

```python
pd.get_dummies(tips_df,drop_first=True) #One hot encoding for every column
```

```python
# Using scikit learn
from sklearn.preprocessing import OneHotEncoder
```

```python
onehotencoder = OneHotEncoder(sparse_output=False)
```

```python
ohe_arr = onehotencoder.fit_transform(tips_df[['sex','smoker','day','time']])
```

```python
ohe_arr #the array is ondehotencoded but the array have more than 4 columns and if we have to create dataframe than we need to get keys from pd.get_dummies
```

```python
# ordinal variables ---> grade,condition,---> USE ORDINALENCODER
# nominal varible --> language,products,colors-->USE LABELENCODER (if data have too many different classes) --->USE ONEHOTENCODER(if column have few classes like male female) 
```

```python
# Ordinal encoding ==>encode data according to level
```

```python
# LabelEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder() # convert data into numbers according to alphabetical order
le.fit_transform(tips_df['day'])
```

```python
le.fit_transform(tips_df['time'])
```

```python
# OrdinalEncoder manuaaly (better approach for many times)
order_label = {'Dinner':2,'Lunch':1} # first, define the order of diffrent classes
tips_df['time'].map(order_label) # then  just map the order on column
```

```python
# Scikit learn OrdinalEncoder ()
from sklearn.preprocessing import OrdinalEncoder
ordinalencoder = OrdinalEncoder()
```

```python
# always keep in mind tnat use ordinalencoder only when the data is in a perticular order
#  I am just applying ordinalencoder for practice 
ordinalencoder.fit_transform(tips_df[['day','time']]) # 2D array required
```
